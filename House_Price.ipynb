```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
#Deep learning models
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.callbacks import EarlyStopping
```

```python
# Dataset
df=pd.read_csv("/content/housing.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
 df.info()
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['ocean_proximity'].value_counts()
```

```python
df['ocean_proximity']=df['ocean_proximity'].map({'<1H OCEAN':0,'INLAND':1,'NEAR OCEAN':2,'NEAR BAY':3,'ISLAND':4})
```

```python
df.head()
```

```python
#Train test split
y=df['median_house_value']
x=df.drop('median_house_value',axis=1)
```

```python
# convert to numpy array
x = np.array(x)
y = np.array(y)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=123)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Scaling
scaler=MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
print(x_train)
print(x_test)
```

```python
#ANN (Building model)

model=Sequential([
    #input layer
    Dense(1000, input_shape=(x_train.shape[1],),activation='relu'),
    Dropout(0.2),
    #Two hidden Layer
    Dense(500,activation='relu'),
    Dropout(0.2),
    Dense(250,activation='relu'),
    Dropout(0.2),
    #output layer
    Dense(1,activation='linear')
])
```

```python
model.summary()
```

```python
#Compile
model.compile(
    optimizer='rmsprop',
    loss='mse',
    metrics=['mae'])
```

```python
#EarlyStopping
es = EarlyStopping(
    monitor='val_loss',
    mode='min',
    patience=50,
    restore_best_weights=True)
```

```python
# fit the model (training)
history = model.fit(x_train, y_train,
                    validation_data= (x_test, y_test),
                    callbacks=[es],
                    epochs=10,
                    batch_size=50,
                    verbose=1)
```

```python
#Metrics and Score

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,mean_squared_log_error

y_pred = model.predict(x_test)
print("mae :", mean_absolute_error(y_test,y_pred))
print("mse :", mean_squared_error(y_test,y_pred))
print("mae :", mean_squared_log_error(y_test,y_pred))
print("score :", r2_score(y_test,y_pred))
```

```python
#Predictive System
def pred(longitude,latitude,housing_median_age,total_rooms,total_bedrooms,population,households,median_income,ocean_proximity):

    features = np.array([longitude,latitude,housing_median_age,total_rooms,total_bedrooms,population,households,median_income,ocean_proximity])

    features_scaled = scaler.fit_transform([features])

    results = model.predict(features_scaled).reshape(1,-1)

    return results[0]
```

```python
longitude = -122.2300
latitude  = 37.8800
housing_median_age  = 41.0000
total_rooms        = 880.0000
total_bedrooms   =  129.0000
population     = 322.0000
households          =    126.0000
median_income       =      8.3252
ocean_proximity       =     3.0000

price = pred(longitude,latitude,housing_median_age,total_rooms,total_bedrooms,population,households,median_income,ocean_proximity)
print(price)
```
